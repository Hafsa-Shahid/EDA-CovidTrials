### Q6. Make 5 questions and solve them by using conditional statements.

```python
# Q1. Find the titles of study that have not yet started recruiting volunteers.

recruit = HS.loc[HS.Status == 'Not yet recruiting']
recruit
```

```python
# Q2. How many trials have their results available?

available_results = HS.loc[HS['Study Results'].isin(['Has Results'])]
available_results
```

```python
# Q3. How many trials are being carried out based on just observations?

trial_count = HS.loc[HS['Study Type'] == 'Observational'].shape[0]
trial_count
```

```python
# Q4. How many case studies are completed and have results?

closed_cases = HS.loc[(HS.Status == 'Completed') & (HS['Study Results'] == 'Has Results')]
closed_cases
```

```python
# Q5. How many trials are completed while being in the first phase of COVID?

initial_stage = HS.loc[(HS.Status == 'Completed') & (HS.Phases == 'Phase 1')]
initial_stage
```
